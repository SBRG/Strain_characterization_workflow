```python
import numpy as np
import scipy as sp
from scipy import sparse
import matplotlib.pyplot as plt

def ReduceExpression(data, centering=False, infoTol=0.8, ZERO=1e-8):
    """ Dataframe organized conditions x gene """    
    M,N = shape(data)
    
    # Perform column mean centering if specified
    data = np.array(data)
    if centering:
        data = [data[:,i] - mean(data[:,i]) for i in xrange(0,N)]    
        data = np.array(data)
        data = data.transpose()        
    
    
    # Perform SVD
    U,s,V = np.linalg.svd(data)   
    m = shape(U)[1]
    n = shape(V)[0]
    nEigen = min(m,n)
    S = sparse.diags(s,0,shape=(m,n))
    A = np.dot(U,(S*V))
    eigengene_range = xrange(1,nEigen)
    N = len(eigengene_range)
    infos = np.zeros(N)
    
    # Remebmer to normalize before svd
    for i,eigengene in enumerate(eigengene_range):
        infos[i] = sum(s[0:eigengene]**2)/sum(s**2)

    nModes = 1 + where(infos >= infoTol)[0][0]
    svsUnion = reduce(lambda i,j: np.union1d( where(abs(V[:,i])>ZERO)[0] , where(abs(V[:,j])>ZERO)[0]), xrange(0,nModes))
    
    # Reproduce the reduced expression matrix with the first k(0.8) modes
    Ared = np.dot(U[:,0:nModes],(S.todense()[0:nModes,0:nModes]*V[0:nModes,:]))    
    Ared[Ared < ZERO] = 0.0    

    # And see what genes are used to explain variation across 180 conditions
    Asum = sum(abs(Ared),axis=0)
    nzinds = Asum.A1.nonzero()

    return Ared,U,S,V,svsUnion,infos
```

```python
def rescale_df(df):
    df_r = (df - df.min())/(df.max()-df.min())
    return df_r
```

```python
def plot_single_SVD(dataframe):

    ### Get major modes for 80% reproduction
    # Column centering would filter out genes with constant expression across conditions--OPPOSITE to what we want
    centering = True # we want to detect variation (true) we want to find the constant values (false)
    infoTol = 0.8

    Ared,U,S,V,svsUnion,info = ReduceExpression(dataframe,centering=centering,infoTol=infoTol)
    first_prin_comp = info[0]
    second_prin_comp = np.subtract(info[1],info[0])

    ### Write the reduced expression matrix
    phenotypic_data = list(dataframe.columns.values)
    time = list(dataframe.index)

    # Conditions
    fig,ax = plt.subplots(figsize=(8,4))

    ymax = 0.2
    ymin = -0.4
    nConds = shape(U)[0]
    dy = (ymax-ymin)/nConds
    ysort = sort(U[:,1]).tolist()

    c='black'
    for i,txt in enumerate(dataframe.index): 
        x = U[i,0] 
        y = U[i,1]    

        r = ysort.index(y)    
        xtxt = U[i,0]+.01
        ytxt = U[i,1]+.01 #ymin+dy*r
        
        if abs(xtxt) > 0.02 and abs(ytxt) > 0.02:
            ax.scatter(x,y,color='k')
            ax.annotate(txt, (x, y), (xtxt,ytxt),color=c, fontsize=10)
    plt.ylabel('Average Second \n Principle Component \n (%f)'%(second_prin_comp), fontsize=15)
    plt.xlabel('Average First Principle Component (%f)'%(first_prin_comp), fontsize=15)
    plt.legend(loc=2)

    return (Ared,U,S,V,svsUnion,info)
    
```
